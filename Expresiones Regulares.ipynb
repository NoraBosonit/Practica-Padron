## Expresiones regulares sencillas

**^\w+\s\w+\s?$**

Le estoy diciendo que me encuentre las expresiones que estén al principio de la línea (^), tengan 1 o más letras seguidas de un espacio, y a continuación 1 o más letras. Luego le estamos indicando que queremos que haya un espacio opcional (?), es decir, que ese espacio (\s) aparezca 0 o 1 veces. Por último ($) indica final de palabra.

Haría matching con expresiones como:

Carlos Arturo (espacio)
Cesar Alan(sin espacio)

Si quiero que me haga matching con nombres simples puedo hacer que la segunda tanda de \w sea opcional. De esta manera:

**^\w+\s?\w+?\s?$**

## Conjuntos de caracteres
Nos permiten buscar caracteres que se encuentres solo dentro de los braquets ([])

Por ejemplo si tenemos números de teléfono tal que:

444 555 66 55

333 777 99 22

Basta con indicar 

**\d{3}\s\d{3}\s\d{2}\s\d{2}**

Sin embargo, no haría matching con este otro:

555-111-44-11

Por esto, podemos indicar que queremos que entre números haya o un espacio o un guión. Esto se hace con los conjuntos de caracteres.

**\d{3}[\s-]\d{3}[\s-]\d{2}[\s-]\d{2}**

Dentro de los conjuntos de caracteres no hay que añadir \ antes de caracteres reservados como el punto.

**[a-zA-Z]+**

Me busca un conjunto de caracteres que contengan letras que se repitan 1 o más veces. Es decir solo palabras

## Grupos
Permitan encerrar trozos de la expresión regular y separarlos en grupos para extraer las coincidencias de una en una

Los grupos sirven para hacer un or o para repetir expresiones.

En el caso de los teléfonos, con los grupos podemos acortar la expresión utilizando grupos:

**\d{3}[\s-]\d{3}([\s-]\d{2}){2}**

O podemos indicar que queremos solo los números que empiecen con 444 o 555:

**(444|555)[\s-]\d{3}([\s-]\d{2}){2}**
